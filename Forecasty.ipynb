# a) Naive Method: 

```python
dd= np.asarray(train.StainlessSteelPrice)
y_hat = test.copy() 
y_hat['naive'] = dd[len(dd)-1] 
plt.figure(figsize=(12,8)) 
plt.plot(train.index, train['StainlessSteelPrice'], label='Train')
plt.plot(test.index,test['StainlessSteelPrice'], label='Test')
plt.plot(y_hat.index,y_hat['naive'], label='Naive Forecast')
plt.legend(loc='best') 
plt.title("Naive Forecast")
plt.show()      
```

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
rms = sqrt(mean_squared_error(test.StainlessSteelPrice, y_hat.naive)) 
print(rms) 
```

# b) Simple Average: 

```python
y_hat_avg = test.copy()
y_hat_avg['avg_forecast'] = train['StainlessSteelPrice'].mean()
plt.figure(figsize=(12,8)) 
plt.plot(train['StainlessSteelPrice'], label='Train') 
plt.plot(test['StainlessSteelPrice'], label='Test') 
plt.plot(y_hat_avg['avg_forecast'], label='Average Forecast') 
plt.legend(loc='best') 
plt.show() 
```

```python
rms = sqrt(mean_squared_error(test.StainlessSteelPrice, y_hat_avg.avg_forecast))
print(rms) 
```

# c) Moving Average:

```python
y_hat_avg = test.copy() 
y_hat_avg['moving_avg_forecast'] = train['StainlessSteelPrice'].rolling(60).mean().iloc[-1]
plt.figure(figsize=(16,8))
plt.plot(train['StainlessSteelPrice'], label='Train')
plt.plot(test['StainlessSteelPrice'], label='Test') 
plt.plot(y_hat_avg['moving_avg_forecast'], label='Moving Average Forecast') 
plt.legend(loc='best') 
plt.show() 
```

```python
rms = sqrt(mean_squared_error(test.StainlessSteelPrice, y_hat_avg.moving_avg_forecast))
print(rms) 
```

# d) Simple Exponential Smoothing 

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt 
y_hat_avg = test.copy()
fit2 = SimpleExpSmoothing(np.asarray(train['StainlessSteelPrice'])).fit(smoothing_level=0.6,optimized=False)
y_hat_avg['SES'] = fit2.forecast(len(test)) 
plt.figure(figsize=(16,8)) 
plt.plot(train['StainlessSteelPrice'], label='Train') 
plt.plot(test['StainlessSteelPrice'], label='Test')
plt.plot(y_hat_avg['SES'], label='SES')
plt.legend(loc='best') 
plt.show() 
```

```python
rms = sqrt(mean_squared_error(test.StainlessSteelPrice, y_hat_avg.SES))
print(rms)
```

# e) Holt-Winters Method: 

```python
y_hat_avg = test.copy()
fit1 = ExponentialSmoothing(np.asarray(train['StainlessSteelPrice']) ,seasonal_periods=7 ,trend='add', seasonal='add',).fit() 
y_hat_avg['Holt_Winter'] = fit1.forecast(len(test))
plt.figure(figsize=(16,8)) 
plt.plot( train['StainlessSteelPrice'], label='Train')
plt.plot(test['StainlessSteelPrice'], label='Test') 
plt.plot(y_hat_avg['Holt_Winter'], label='Holt_Winter') 
plt.legend(loc='best')
plt.show() 
```

```python
rms = sqrt(mean_squared_error(test.StainlessSteelPrice, y_hat_avg.Holt_Winter)) 
print(rms) 
```

# 4. Analysis

### Results of Regression model:
##### In the previous steps, 2 multi-linear regression, and 3 simple linear regression models were tested to find apprpriate features for forecasting. The criteria for selecting proper features are MSE, RMSE, and MAE  . In the first model all features are considered. In the second model 5 features were dropped. These features have high correlation with the other features in distinct pairs, and in order to test the impact of multicolineraity, they are dropped. Finally, top 3 features which have the highest correlation with dependent variable (y) are tested separatly in three sinple regression models. Among 5 different tests that are done to find the best combination of features in the linear regression models, the model with a single independent variable of "Nickel_Global_USD" has the least error value (MSE equals 0.006). Therefore, the simple regression model with  "Nickel_Global_USD" as an independent variable performs better in these tests. However, by implementing the Forward/ Beackward Greedy Heuristic we can find best combination with the least error metric value. 

### Results of Naive model:
#### As it's obvious from the various implemented Naive models based on their graphical models, they are not much reliable. So the regressions are better for forecasting of the steel price.
